```python
import pandas as pd
import json

# 1. Load the data
# Use the exact filename of your CSV
df = pd.read_csv('DOB_Safety_Violations.csv')

# 2. Convert 'Violation Issue Date' to datetime
# This column is Column B in your screenshot
df['Violation Issue Date'] = pd.to_datetime(df['Violation Issue Date'], errors='coerce')

# 3. Filter for records starting Jan 1, 2026
df_2026 = df[df['Violation Issue Date'] >= '2026-01-01'].copy()

# 4. Group by BIN to create the Risk Register
# Key = BIN (Column A), Value = List of risks
risk_register = {}

for bin_id, group in df_2026.groupby('BIN'):
    risk_register[str(bin_id)] = group[[
        'Violation Issue Date',
        'Violation Type',
        'Violation Remarks',
        'Violation Status'
    ]].to_dict('records')

# 5. Save as the 'Gold' JSON file
with open('cleaned_risk_register.json', 'w') as f:
    # default=str handles the Timestamp objects in the dictionary
    json.dump(risk_register, f, indent=4, default=str)

print(f"Success! Created risk registers for {len(risk_register)} unique buildings.")
```

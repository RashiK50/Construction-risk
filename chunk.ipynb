```python
import pandas as pd
import json
```

```python
df = pd.read_csv("DOB_NOW_cleaned.csv")
```

```python
sheet_1_cols = [

    "job_filing_number",
    "job_type",
    "borough",
    "block",
    "lot",
    "bin",
    "building_type",
    "initial_cost",
    "total_construction_floor_area",
    "existing_stories",
    "proposed_no_of_stories",
    "existing_dwelling_units",
    "proposed_dwelling_units",
    "filing_date",
    "approved_date",
    "postcode",
    "latitude",
    "longitude"
]

sheet_2_cols = [

    "applicant_professional_title",
    "applicant_license_",
    "applicant_first_name",
    "applicant_last_name",

    "owner_s_business_name",

    "filing_representative_business_name",

    "specialinspectionrequirement",
    "progressinspectionrequirement",

    "review_building_code",

    "boiler_equipment_work_type_",
    "earth_work_work_type_",
    "foundation_work_type_",
    "general_construction_work_type_",
    "mechanical_systems_work_type_",
    "place_of_assembly_work_type_",
    "protection_mechanical_methods_work_type_",
    "sidewalk_shed_work_type_",
    "structural_work_type_",
    "support_of_excavation_work_type_",
    "temporary_place_of_assembly_work_type_",

    "standpipe",
    "antenna",
    "curb_cut",
    "sign",
    "fence",
    "scaffold",
    "shed"
]
```

```python
df = df.fillna("UNKNOWN")
```

```python
project_profiles = []

for _, row in df.iterrows():

    profile = {

        "job_number": str(
            row.get("job_filing_number", "UNKNOWN")
        ),

        "sheet_1_project_details": {

            col: str(row[col])

            for col in sheet_1_cols
        },

        "sheet_2_contractor_scope": {

            col: str(row[col])

            for col in sheet_2_cols
        }
    }

    project_profiles.append(profile)
```

```python
with open(
    "structured_project_profiles.json",
    "w"
) as f:

    json.dump(
        project_profiles,
        f,
        indent=2
    )

print("Structured project profiles saved!")
```

```python
documents = []

for profile in project_profiles:

    s1 = profile["sheet_1_project_details"]
    s2 = profile["sheet_2_contractor_scope"]

    text = f"""

    PROJECT DETAILS:

    Job Type: {s1.get('job_type')}
    Borough: {s1.get('borough')}
    BIN: {s1.get('bin')}
    Block: {s1.get('block')}
    Lot: {s1.get('lot')}

    Building Type: {s1.get('building_type')}

    Initial Cost: {s1.get('initial_cost')}

    Floor Area:
    {s1.get('total_construction_floor_area')}

    Existing Stories:
    {s1.get('existing_stories')}

    Proposed Stories:
    {s1.get('proposed_no_of_stories')}

    Existing Dwelling Units:
    {s1.get('existing_dwelling_units')}

    Proposed Dwelling Units:
    {s1.get('proposed_dwelling_units')}

    Filing Date:
    {s1.get('filing_date')}

    CONTRACTOR & SCOPE:

    Applicant Professional Title:
    {s2.get('applicant_professional_title')}

    Owner Business:
    {s2.get('owner_s_business_name')}

    Filing Representative:
    {s2.get('filing_representative_business_name')}

    Building Code:
    {s2.get('review_building_code')}

    Structural Work:
    {s2.get('structural_work_type_')}

    Foundation Work:
    {s2.get('foundation_work_type_')}

    Mechanical Systems Work:
    {s2.get('mechanical_systems_work_type_')}

    General Construction Work:
    {s2.get('general_construction_work_type_')}

    Support Of Excavation:
    {s2.get('support_of_excavation_work_type_')}

    Special Inspection Requirement:
    {s2.get('specialinspectionrequirement')}

    Progress Inspection Requirement:
    {s2.get('progressinspectionrequirement')}

    Scaffold:
    {s2.get('scaffold')}

    Shed:
    {s2.get('shed')}
    """

    documents.append(text)
```

```python
with open(
    "embedding_documents.json",
    "w"
) as f:

    json.dump(
        documents,
        f,
        indent=2
    )

print("Semantic chunks created successfully!")
```
